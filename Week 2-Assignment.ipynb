```python
#import modules

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# import data

original_train=pd.read_csv('train.csv', index_col=0)
original_test=pd.read_csv('test.csv',index_col=0)
```

```python
#find empty data
mostly_empty_features = []
for col in original_train.columns:
    if original_train[col].count()<0.5*original_train[col].size:
        mostly_empty_features.append(col)
        
print('The following features has 50% or more empty cells:', mostly_empty_features)
```

```python
#Combining train and test data to fill NA values in both sets together
original_test['SalePrice']=0
original_test['Source']=1
original_train['Source']=0

original_combined=pd.concat([original_train,original_test])
```

```python
#data preparation

# filling up NA values based on the data description

#Another way is to do a for loop: None_features=['Alley','BsmtQual'...]
#for col in None_features:
    #original_combined ['Col']=original_combined['Col'].fillna('None')

original_combined ['Alley']=original_combined['Alley'].fillna('None')
original_combined ['BsmtQual']=original_combined['BsmtQual'].fillna('None')
original_combined ['BsmtCond']=original_combined['BsmtCond'].fillna('No Basement')
original_combined ['BsmtExposure']=original_combined['BsmtExposure'].fillna('None')
original_combined ['BsmtFinType1']=original_combined['BsmtFinType1'].fillna('No Basement')
original_combined ['BsmtFinType2']=original_combined['BsmtFinType2'].fillna('No Basement')
original_combined ['FireplaceQu']=original_combined['FireplaceQu'].fillna('No Fireplace')
original_combined ['GarageType']=original_combined['GarageType'].fillna('No Garage')
original_combined ['GarageCond']=original_combined['GarageCond'].fillna('No Garage')
original_combined ['PoolQC']=original_combined['PoolQC'].fillna('No Pool')
original_combined ['Fence']=original_combined['Fence'].fillna('No Fence')
original_combined ['MiscFeature']=original_combined['MiscFeature'].fillna('None')
original_combined ['GarageQual']=original_combined['GarageQual'].fillna('No Garage')
original_combined ['GarageFinish']=original_combined['GarageFinish'].fillna('No Garage')
original_combined ['Exterior2nd']=original_combined['Exterior2nd'].fillna('No Exterior2nd')
original_combined ['MasVnrType']=original_combined['MasVnrType'].fillna('None')
original_combined ['MasVnrArea']=original_combined['MasVnrArea'].fillna(0)
original_combined ['BsmtFinSF1']=original_combined['BsmtFinSF1'].fillna(0)
original_combined ['BsmtFinSF2']=original_combined['BsmtFinSF2'].fillna(0)
original_combined ['TotalBsmtSF']=original_combined['TotalBsmtSF'].fillna(0)
original_combined ['BsmtUnfSF']=original_combined['BsmtUnfSF'].fillna(0)
original_combined ['BsmtFullBath']=original_combined['BsmtFullBath'].fillna(0)
original_combined ['BsmtHalfBath']=original_combined['BsmtHalfBath'].fillna(0)
original_combined ['GarageYrBlt']=original_combined['GarageYrBlt'].fillna(0)
original_combined ['GarageCars']=original_combined['GarageCars'].fillna(0)
original_combined ['GarageArea']=original_combined['GarageArea'].fillna(0)


#fill all na values for the remanining features, the missing value is difficut to gage so will fill it with the preceeding value
original_combined=original_combined.fillna(method='ffill')
original_combined=original_combined.fillna(method='bfill')


#some_empty_features = []
#for col in original_combined.columns:
    #if original_combined[col].count()<original_combined[col].size:
        #some_empty_features.append(col)
        
#print('The following features has empty cells:', some_empty_features)



#Numerical categorical values are changed to category

for col in original_combined.columns:
    if original_combined[col].dtypes=='object' or col=='MSSubClass':
        original_combined[col]=original_combined[col].astype('category')

original_combined.info()

#categorical_features = ['MSSubClass']
```

```python
#Get dummy variables for all categorical values
original_combined=pd.get_dummies(original_combined)

#train and test sets

training_set = original_combined[original_combined['Source']==0]
validation_set = original_combined[original_combined['Source']==1]
validation_set = validation_set.drop(columns='SalePrice')
```

```python
def corr_chart(df_corr):
    corr=df_corr.corr()
    
    #screen top half to get a triangle
    
    top = np.zeros_like(corr, dtype=np.bool)
    top[np.triu_indices_from(top)] = True
    fig=plt.figure()
    fig, ax = plt.subplots(figsize=(12,12))
    sns.heatmap(corr, mask=top, cmap='coolwarm', 
        center = 0, square=True, 
        linewidths=.5, cbar_kws={'shrink':.5}, 
        annot = True, annot_kws={'size': 9}, fmt = '.3f')           
    plt.xticks(rotation=45) # rotate variable labels on columns (x axis)
    plt.yticks(rotation=0) # use horizontal variable labels on rows (y axis)
    plt.title('House Features Correlation Heat Map')   
    plt.savefig('House Features Correlation map.pdf', bbox_inches = 'tight', dpi=None, facecolor='w', edgecolor='b', 
        orientation='portrait', papertype=None, format=None,transparent=True, pad_inches=0.25)
```

```python
#import regression models
import sklearn
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

#import scaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
x = original_combined.drop(columns='SalePrice')
y = original_combined['SalePrice']
```

```python
#crossvalidation by one split once 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

#Exploring 10-fold Cross Val in Linear Regression
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

Newx = training_set.drop(columns='SalePrice')
Newy = training_set['SalePrice']


x_train, x_test, y_train, y_test = train_test_split(Newx,Newy, test_size=0.2, random_state=0)

```

```python
#Exploring Linear Regression with StandardScaler 
scaler=StandardScaler()
x_train_scaled =scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

model=LinearRegression()
model.fit(x_train_scaled, y_train)
y_pred = model.predict(x_test_scaled)

RMSE =np.sqrt(mean_squared_error(y_test,y_pred))
plt.xlabel('Actual Sales Price')
plt.ylabel('Predicted Sales Price')

print(RMSE)

    
plt.scatter(y_test,y_pred, color='r')
plt.grid(linestyle=':')
#plt.show()

y1_test=np.array(y_test).reshape(-1,1)
y1_pred=y_pred.reshape(-1,1)


linreg=LinearRegression()
linreg.fit(y1_test,y1_pred)
y1_fitted=linreg.coef_*y1_test+linreg.intercept_

text="Slope={}.".format(linreg.coef_)
plt.plot(y1_test,y1_fitted, color='g', linestyle='-',linewidth=1)
plt.text(100000,50000000000000000,text)
plt.title('Linear Regression with StandardScaler')   
plt.savefig('Linear Regression with StandardScaler.pdf')

plt.show()

```
